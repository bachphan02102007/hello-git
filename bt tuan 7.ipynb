```python
#W7A1 
def binary_search(arr, target):
    left=0
    right=len(arr)-1
    while left<=right:
        mid=(left+right)//2
        if arr[mid]==target:
            return mid
        elif arr[mid]<target:
            left=mid+1
        else:
            right=mid-1
    if left>right:
        return -1
    else:
        return mid
arr=list(map(int,input().split()))
target=int(input())
print(binary_search(arr,target))
      
```

```python
#W7A2
def count_occurrences(a, x):
    dem=0
    for i in a:
        if i==x:
            dem+=1
    return dem       
a=list(map(int,input().split()))
x=int(input())
print(count_occurrences(a,x))
```

```python
#W7A3
def bubble_sort(a):
    for i in range(len(a)):
        swap=False
        for j in range(0,len(a)-i-1):
            if a[j]>a[j+1]:
                a[j],a[j+1]=a[j+1],a[j]
                swap=True
        if not swap:
            break
    return a
a=list(map(int,input().split()))
print(bubble_sort(a))
```

```python
#W7A3
def selection_sort(a):
    for i in range(len(a)):
        minn=i
        for j in range(i+1,len(a)):
            if a[j]<a[minn]:
                a[j],a[minn]=a[minn],a[j]
    return(a)
a=list(map(int,input().split()))
print(selection_sort(a))
```

```python
#W7A4
def xh_max(a):
    d={}
    for i in a:
        if i not in d:
            d[i]=1
        else:
            d[i]+=1
    maxx=max(d.values())
    for key in d:
        if d[key]==maxx:
            return (f"{key} xuat hien nhieu nhat, som nhat, {maxx} lan")

a=list(map(int,input().split()))
print(xh_max(a))
```

```python
#W7A5
def so_cap(a,x):
    dem=0
    for i in range(len(a)):
        for j in range(i+1,len(a)):
            if (a[i]+a[j])==x:
                dem+=1
    return dem
a=list(map(int,input().split()))
x=int(input())
print(so_cap(a,x))
```

```python
#W7A9
def phan_tu_nho_hon(a):
    c=[]
    for i in range(len(a)):
        dem=0
        for j in range(i+1,len(a)):
            if a[i]>a[j]:
                dem+=1
        c.append(dem)    
    return c   
a=list(map(int,input().split()))
print(phan_tu_nho_hon(a))
```

```python
#W7A13
def hai_hoa(a):
    a.sort()
    c=[]
    for i in range(len(a)):
        dem=1
        for u in range(i+1,len(a)):
            if a[u]-a[i]<=1:
                dem+=1
            else:
                break
        c.append(dem)
    return max(c)
a=list(map(int,input().split()))
print(hai_hoa(a))
```
