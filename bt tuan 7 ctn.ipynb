```python
#W7A16 Cho một mảng số nguyên không âm. Gọi
# mảng đó là "mảng đẹp" nếu mỗi phần tử
# x trong mảng xuất hiện đúng x lần.
# Kiểm tra xem mảng có phải là mảng đẹp
# hay không. Nếu không hãy xóa đi y số (y
# là nhỏ nhất có thể) để mảng còn lại thành
# mảng đẹp.
def mang_dep(a):
    a.sort()
    b=set(a)
    soso=0
    for i in b:
        dem=0
        for u in a:
            if u==i:
                dem+=1
            if u>i:
                break
        if dem>i:
            soso+=(dem-i)
        elif dem<i:
            soso+=dem
        else:
            continue
    if soso==0:
        return 'YES'
    return f"NO {soso}"
a=list(map(int,input().split()))
print(mang_dep(a))
```

```python
#W7A20 Cho dãy N số nguyên dương, tính tổng
# các số có k chữ số. Dòng đầu ghi N số
# nguyên dương, dòng sau ghi số nguyên
# dương k
def tong(a,n):
    a.sort()
    t=0
    for i in a:
        if 10**(n-1)<=i<10**n:
            t+=i
        elif i>10**n:
            break
    return t
a=list(map(int,input().split()))
n=int(input())
print(tong(a,n))
```

```python
#W7A24 Cho xâu kí tự, xóa các từ giống nhau liền kề, chỉ giữ lại từ đầu tiên
def xoa_ky_tu_trung(a):
    b=[]
    for i in range(1,len(a)):
        if a[i]==a[i-1]:
            b.append(i)
    b.reverse()
    for i in b:
        del a[i]
    s=" ".join(a)
    return s
a=list(input().split())
print(xoa_ky_tu_trung(a))
```

```python
#W7A28Bà̀i toán Tìm dãy các hợp số dài nhất liên
# tiếp không chứa số nguyên nào <= N (N
# là số nguyên dương cho trước). Nếu có
# nhiều dãy cùng độ dài thì lấy dãy nhỏ
# nhất (gặp đầu tiên từ trái qua)
import math
def lien_hop_min(a):
    b=[]
    for i in range(2,a+1):
        for u in range(2,math.isqrt(i)+1):
            if i%u==0:
                break
        else:
            b.append(i)
    b.reverse()
    e=0
    d=0
    maxx=0
    for i in range(len(b)-1):
        c= b[i]-b[i+1]
        if c>=maxx:
            maxx=c
            e=b[i]
            d=b[i+1]
    f=[]
    for i in range(d+1,e,1):
        f.append(i)
    return f
a=int(input())
print(lien_hop_min(a))
```

```python
#W7A28Bà̀i toán Tìm dãy các hợp số dài nhất liên
# tiếp không chứa số nguyên nào <= N (N
# là số nguyên dương cho trước). Nếu có
# nhiều dãy cùng độ dài thì lấy dãy nhỏ
# nhất (gặp đầu tiên từ trái qua)
import math
def lien_hop_min(a):
    b=[]
    for i in range(2,a+1):
        for u in range(2,math.isqrt(i)+1):
            if i%u==0:
                break
        else:
            b.append(i)
    b.reverse()
    maxx=b[0]-b[1]
    cap=(b[0],b[1])
    for i in range(1,len(b)-1):
        c= b[i]-b[i+1]
        if c>maxx:
            cap=(b[i],b[i+1])
    f=[]
    for i in range(cap[1]+1,cap[0],1):
        f.append(i)
    return f
a=int(input())
print(lien_hop_min(a))
```

```python
#W7A29 Tìm Dãy con tăng dần liên tục dài nhất
# của một danh sách nhập vào từ bàn
# phím. Lấy dãy đầu tiên nếu có nhiều dãy
# có số phần tử bằng nhau
def con_tang(a):
    maxx=0
    e=0
    d=0
    for u in range(len(a)-1):
        dem=0
        for i in range(u,len(a)-1):
            if a[i]<a[i+1]:
                dem+=1
            else:
                break
        if dem>maxx:
            maxx=dem 
            e=u
            d=i
        elif dem==maxx:
            continue
    j=[]
    for i in range(e,d+1):
        j.append(a[i])
    return j
a=list(map(int,input().split()))
print(con_tang(a))
```

```python
def con_tang(a):
    max_len = 1
    start = 0
    cur_len = 1
    cur_start = 0
    
    for i in range(1, len(a)):
        if a[i] > a[i-1]:
            cur_len += 1
        else:
            if cur_len > max_len:
                max_len = cur_len
                start = cur_start
            cur_start = i
            cur_len = 1
    
    # kiểm tra đoạn cuối
    if cur_len > max_len:
        max_len = cur_len
        start = cur_start
    
    return a[start:start+max_len]

a = list(map(int, input().split()))
print(con_tang(a))
```

```python
#W7A30 Tìm Dãy con tăng dần KHÔNG liên tục
# dài nhất của một danh sách nhập vào từ
# bàn phím 
def con_tang_klt(a):
    c=[]
    for i in range(len(a)):
        b=[]
        b.append(a[i])
        for u in range(i+1,len(a)):
            if u>=b[len(b)-1]:
                b.append(a[u])
        c.append(b)
    maxx=0
    for i in c:
        if len(i)>maxx:
            maxx=len(i)
    for u in c:
        if len(u)==maxx:
            return u
a=list(map(int,input().split()))
print(con_tang_klt(a))
```
