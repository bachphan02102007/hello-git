```python
#W9A1 Viết hàm merge hai danh sách đã sắp xếp mà không dùng bất cứ thuật toán sắp xếp nào 
def merge(a,b):
    n=[]
    them_x=False
    x=a.pop(0)
    y=b.pop(0)
    if x<=y:
        n.append(x)
        them_x=True
    else:
        n.append(y)
        them_x=False
    if them_x is True :
        b.insert(0,y)
    else:
        a.insert(0,x)
    if not a :
        n=n+b
        return n
    if not b:
        n=n+a
        return n
    merge(a,b)
a=list(map(int,input().split()))
b=list(map(int,input().split()))
print(merge(a,b))
```
