```python
#W6A3
def xoay(a,k):
    k=k%len(a)
    a=a[k:]+a[:k]
    return a
a=tuple(map(int,input().split()))
k=int(input())
print(xoay(a,k))
```

```python
#W6A4
def dic(a):
    d={}
    for i in a:
        key,value=i.split(':')
        if key not in d:
            d[key]=[value]
        else:
            d[key].append(value)
    return d
a=input().split()
print(dic(a))
```

```python
#W6A5
def phan_loai(a):
    d={'pst':0,'zeros':0,'ngt':0}
    for i in a:
        if i>0:
            d['pst']+=1
        elif i==0:
            d['zeros']+=1
        else :
            d['ngt']+=1
    return d
a=list(map(int,input().split()))
print(phan_loai(a))
```

```python
#W6A9
def tong_dic(a,b):
    c={}
    for i in a:
        key,value=i.split(':')
        if key not in c:
            c[key]=[value]
        else:
            c[key].append(value)
    for u in b :
        key,value=u.split(':')
        if key not in c:
            c[key]=[value]
        else :
            c[key].append(value)
    return c
a=input().split()
b=input().split()
print(tong_dic(a,b))
```

```python
#W6A10
def cap(a,k):
    m=[]
    for i in range((len(a))):
        for u in range(i+1,len(a)):
            if a[i]+a[u]==k:
                m.append((a[i],a[u]))
    return m
a=list(map(int,input().split()))
k=int(input())
print(cap(a,k))
```

```python
#W6A12
def xh(a):
    d={}
    c=[]
    for i in a:
        if i not in d:
            d[i]=1
        else:
            d[i]+=1
    m=max(d.values())
    for key in d:
        if d[key]==m:
            c.append(key)
    return min(c)
a=list(map(int,input().split()))
print(xh(a))
```

```python
#W6A14
def phan_chung(a,b):
    c=[]
    for i in b:
        if i in a:
            c.append(i)
    return c
a=list(map(int,input().split()))
b=list(map(int,input().split()))
print(phan_chung(a,b))
```

```python
#W6A15
def vl(a,k):
    d={}
    for i in a:
        key,value=i.split(':')
        value=int(value)
        if value >k:
            d[key]=[value]
    return d
a=input().split()
k=int(input())
print(vl(a,k))
```
