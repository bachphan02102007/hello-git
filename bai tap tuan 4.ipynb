```python
#W4A6
while True :
    try:
        a,b=map(int,input().split())
        if a>b or a<=0 or b<=0:
            print('nhap lai')
            continue
        break
    except ValueError:
        print('sai du lieu')
import math
tong=0
for i in range(a,b+1):
    if i ==1:
        continue
    for u in range(2,math.isqrt(i)+1):
        if i%u==0:
            break
    else:
        tong+=i
print(tong)
```
