```python
#W6A16 Nhâ ̣p vào mảng n hàng, m cột số nguyên. In lại mảng ra màn hình như ma ̣
# trận hai chiều, định dạng mỗi phần tử 4 vị trí căn phải

def matrix(m,n):
    a=[]
    for i in range(m):
        a.append(list(map(int,input().split())))
    for i in range(m):
        for j in range(n):
            print(f"{a[i][j]:>2d}",end=" ")
        print()
m,n=map(int,input().split())
matrix(m,n)
```

```python
#W6A18Nhâp vào mảng n hàng, m cột số nguyên và số nguyên k (n, m, k nhập trên ̣
# cùng một dòng, cách nhau dấu cách). Tính và in ra tổng các phần tử trên cột k (hàng k) 
def tong_hang_cot_bat_ky(m,n,k):
    a=[]
    b=[]
    c=[]
    for i in range(m):
        a.append(list(map(int,input().split())))
    for i in range(m):
        for j in range(n):
            if j==k-1:
                b.append(a[i][j])
            if i==k-1:
                c.append(a[i][j])
    print(sum(b))
    print(sum(c))
m,n,k=map(int,input().split())
tong_hang_cot_bat_ky(m,n,k)
```

```python
#W6A19 Viết hàm đưa vào 1 list số nguyên, tìm và trả về vị trí có giá trị lớn nhất trong list
def maxx(a):
    m=0
    for i in a:
        if i>=m:
            m=i
    return m
a=list(map(int,input().split()))
print(maxx(a))
```

```python
#W6A20 Viết hàm đưa vào một list số nguyên và một số nguyên dương k. Hãy tìm và 
# trả về vị trí của phần tử đầu tiên có giá trị k trong list, nếu không có thì trả về -1
def tim_so(a,k):
    if k not in a:
        print(-1)
    else:
        vitri=0
        for i in a:
            vitri+=1
            if i==k:
                break
        print(vitri)
a=list(map(int,input().split()))
k=int(input())
tim_so(a,k)
```

```python
#W6A21 Nhập một dòng gồm các chuỗi (tách bởi dấu cách). Tạo dictionary ánh xạ 
# chuỗi -> tuple các vị trí (0-based) mà chuỗi đó xuất hiện
a=input().split()
d={}
for i in range(len(a)):
    if a[i] not in d:
        d[a[i]]=[i]
    else:
        d[a[i]].append(i)
for key in d:
    d[key]=tuple(d[key])
print(d)
```

```python
#W6A23 VIết hàm xóa đi các số <x có trong mảng
def xoa(a,x):
    b=[]
    for i in a:
        if i>x:
            b.append(i)
    print(b)
a=list(map(int,input().split()))
x=int(input())
xoa(a,x)
```

```python
#W6A24 Viêết hàm đẩy tất cả các số 0 về đầu (về cuối), các số khác vẫn đúng vị trí
# trước sau: 0 2 3 2 4 0 0 0 9 => 0 0 0 2 3 2 4 9
def day(a):
    n=0
    for i in a:
        if i==0:
            n+=1
    while 0 in a:
        a.remove(0)
    b=n*[0]+a
    c=a+n*[0]
    print(b,c,sep='\n')
a=list(map(int,input().split()))
day(a)
```

```python
#W6A26 Tính tổng các phần tử cực đại địa phương trong một danh sách, không tính 2 phần tử đầu và cuối
a=list(map(int,input().split()))
tong=0
for i in range(1,len(a)-1):
    if a[i]>a[i-1] and a[i]>a[i+1]:
        tong+=a[i]
print(tong)
```
