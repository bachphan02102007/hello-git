```python
#W8A1 [Cylinder}
# Xây dựng lớp Cylinder (hình trụ) gồm: bán kính và
# chiều cao (kiểu số thực) với hai phương thức:tính
# diện tích bề mặt (diện tích toàn phần) và thể tích
# của hình trụ
# Đầu vào
# Đâu vào từ bàn phím gồm một dòng duy nhất
# chứa bán kính và chiều cao của hình trụ, cách
# nhau bởi một dấu cách.
# Đầu ra
# In ra màn hình diện tích bề mặt và thể tích của
# hình trụ (sử dụng lại những hàm vừa viết, π=3.
# 14).
# (Kết quả làm tròn đến 2 chữ số sau dấu phẩy).
class Cylinder:
    def __init__(self,r,h):
        self.r=r
        self.h=h
    def s_tp(self):
        return self.r**2*3.14*2+self.r*2*3.14*self.h
    def tt(self):
        return self.r**2*3.14*self.h
r,h=map(int,input().split())
s=Cylinder(r,h)
print(f"dien tich be mat la:{s.s_tp():.2f}")
print(f"the tich la:{s.tt():.2f}")
```

```python
#W8A2 [Date]
# Xây dựng lớp Date biểu diễn ngày tháng năm với
# hai phương thức kiểm tra xem một ngày tháng
# nào đó có hợp lệ hay không và phương thức tính
# ngày kế tiếp của ngày tháng năm nào đó nếu nó là
# hợp lệ
# Nhập vào một ngày tháng năm, (ngày tháng năm
# biểu diễn dưới dạng dd/mm/yyyy). Nếu ngày
# tháng năm không tồn tại, in ra dòng chữ INVALID.
# Nếu là hợp lệ in ra ngày kế tiếp của nó.
# Đầu vào
# Đầu vào từ bàn phim chứa dữ liệu ngày, tháng,
# năm dưới dạng dd/mm/yyyy .
# Đầu ra
# In ra màn hình ngày kế tiếp của ngày vừa nhập
# dưới dạng dd/mm/yyyy
# . Nếu ngày tháng không tồn tại, in ra INVALID
class Date:
    def __init__(self,d,m,y):
        self.d=d
        self.m=m
        self.y=y
    def Nam_nhuan(self):
        return (self.y % 400 == 0) or (self.y % 4 == 0 and self.y % 100 != 0)
    def Is_valid(self,b):
        if Date.Nam_nhuan(self):
            b[2]=29
        if 0<self.m<=12 and 0<self.d<=b[self.m]:
            return True
        return False
    def Next_day(self,b):
        if Date.Nam_nhuan(self):
            b[2]=29
        if self.d<b[self.m]:
            return f"{(self.d+1):02d}/{self.m:02d}/{self.y:02d}"
        elif self.m<12:
            return f"01/{(self.m+1):02d}/{self.y:02d}" 
        else:
            return f"01/01/{(self.y+1):02d}"
d,m,y=map(int,input().split('/'))
s=Date(d,m,y)    
b=[0,31,28,31,30,31,30,31,31,30,31,30,31]
if not s.Is_valid(b):
    print("INVALID")
else:
    print(s.Next_day(b))
```

```python
#W8A3 [Point]
# Xây dựng lớp điểm với hai thuộc tính x, y cùng
# các phương thức kiểm tra xem có phải gốc tọa dộ
# không, có nằm trên trục tung không, có nằm trên
# trục hoành không, tính khoảng cách từ điểm đó tới
# một điểm khác
# Đầu vào:
# Nhập vào tọa độ một điểm M bất kỳ
# Đầu ra:
# Thông báo M có phải gốc tọa độ không hoặc có
# nằm trên trục tung hoặc trục hoành không, khoảng
# cách từ M tới gốc tọa độ là bao nhiêu (lấy 2 chữ
# số phần thập phân)
class Point:
    def __init__(self,x,y):
        self.x=x
        self.y=y
    def is_goc(self):
        return self.x==0 and self.y==0
    def in_th(self):
        return self.y==0
    def in_tt(self):
        return self.x==0
    def distance(self):
        return f"{((self.x**2+self.y**2)**(1/2)):.2f}"
x,y=map(int,input().split())
s=Point(x,y)
if s.is_goc():
    print(f"M LA GOC TOA DO")
elif s.in_th():
    print(f"M NAM TREN TRUC HOANH:{s.distance()}")
elif s.in_tt():
    print(f"M NAM TREN TRUC TUNG:{s.distance()}")
else:
    print(f"M KHONG THUOC TRUC NAO:{s.distance()}")

    
```

```python
#W8A4 [Calculator]
# Xây dựng lớp Calculator (máy tính bỏ túi) với các
# phương thức để thực hiện các phép toán 1. add
# (cộng), 2. subtract(trừ), 3. product (nhân) và 4.
# divide(chia), 5. power(hàm mũ), 6. mod (lấy phần
# dư) và phương thức 7. set_numbers(cập nhật 2
# số mới) nếu người dùng muốn tính toán trên 2 số
# khác.
# Khởi tạo lớp là hai thuộc tính gồm 2 số cần tính,
# Mỗi phương thức thực hiện một phép tính và trả
# về kết quả tương ứng
# Tạo menu và cho người dùng tính toán ta,tới khi gõ
# "yes", "YES" hoặc "Yes" cho câu hỏi "Bạn có
# muốn thoát?" mới dừng.
class Calculator:
    def __init__(self,a,b):
        self.a=a
        self.b=b
    def add(self):
        return f"{self.a}+{self.b}={self.a+self.b}"
    def subtract(self):
        return f"{self.a}-{self.b}={self.a-self.b}"
    def product(self):
        return f"{self.a}*{self.b}={self.a*self.b}"
    def divide(self):
        return f"{self.a}/{self.b}={self.a/self.b}"
    def power(self):
        return f"{self.a}^{self.b}={self.a**self.b}"
    def mod(self):
        return f"{self.a}%{self.b}={self.a%self.b}"
    def set_number(self,c,d):
        self.a=c
        self.b=d
a,b=map(int,input().split())
s=Calculator(a,b)
for _ in range(1000):
    print("\nMENU")
    print("1. Cộng")
    print("2. Trừ")
    print("3. Nhân")
    print("4. Chia")
    print("5. Mũ")
    print("6. Mod")
    print("7. Đổi 2 số")
    x=input('chon')
    if x=='1':
        print(s.add())
    elif x=='2':
        print(s.subtract())
    elif x=='3':
        print(s.product())
    elif x=='4':
        print(s.divide())
    elif x=='5':
        print(s.power())
    elif x=='6':
        print(s.mod())
    else:
        c,d=map(int,input('NHAP 2 SO MOI').split())
        s.set_number(c,d)
    x=input('BAN CO MUON THOAT')
    if x=='yes' or x=='YES'or x=='Yes':
        break

        
        
```
