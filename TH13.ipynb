```python
#W11A4
from collections import Counter as ct
def areAnagrams(path1, path2):
    with open(path1) as f1:
        n1 = f1.read().strip()
    with open(path2) as f2:
        n2 = f2.read().strip()
    return ct(n2) == ct(n1)

areAnagrams("TH13-4-1.txt", "TH13-4-2.txt")
```

```python
#W11A10
def getMoney(path):
    with open(path) as f:
        money = f.read().strip().split()
        return [int(x) for x in money]
    
getMoney("TH13-10.txt")


def rob(houses):
    if houses == []:
        return 0
    if len(houses) == 1:
        return houses[0]
    
    dp = [0] * len(houses)
    dp[0] = houses[0]
    dp[1] = max(houses[0], houses[1])

    for i in range(2, len(houses)):
        dp[i] = max(dp[i-1], dp[i-2] + houses[i])

    return dp[-1]

rob(getMoney("TH13-10.txt"))
    
```

```python
#W11A12
def maximumProduct(path):
    with open("TH13-12.txt") as f:
        n = int(f.readline().strip())
        num = [int(x) for x in f.read().split()]
    num.sort()
    x1 = num[-1] * num[-2] * num[-3]
    x2 = num[0] * num[1] * num[-1]
    return max(x1, x2)

maximumProduct("TH13-12.txt")
```

```python
#W11A13
def nozero(path):
    with open(path) as f:
        n = int(f.readline().strip())
        num = [int(x) for x in f.read().split()]

    not_zero = [x for x in num if x != 0]
    zero_count = num.count(0)

    return not_zero + [0] * zero_count

nozero("TH13-13.txt")
```

```python
#W11A14
def averageTime(path):
    with open(path) as file:
        lines = file.readlines()[1:] 
        times = [float(line.strip().split(',')[1]) for line in lines]
    
    average = sum(times) / len(times)
    return round(average, 2)

averageTime("TH13-14.csv")
```

```python
#W11A15
def findMovies(path):
    result = []
    with open(path) as file:
        lines = file.readlines()[1:]

        for line in lines:
            parts = line.strip().split(',')
            if len(parts) < 4:
                continue

            movie = parts[1].strip()
            description = parts[2].strip().lower()
            rating = float(parts[3].strip())

            if rating >= 8.0 and "boring" not in description:
                result.append(movie)

    return result

findMovies("TH13-15.csv")
```

```python
#W11A16
import math

def computeAreas(path):
    with open(path) as f:
        lines = [line.strip() for line in f if line.strip()]
    
    for i in range(0, len(lines), 2):
        shape = lines[i].split(':')[1].strip().upper()
        data = lines[i+1]

        if shape == 'SQUARE':
            side = float(data.split(':')[1])
            area = side ** 2
        elif shape == 'RECTANGLE':
            w = float(data.split(':')[1].split()[0])
            h = float(data.split('HEIGHT:')[1])
            area = w * h
        elif shape == 'CIRCLE':
            r = float(data.split(':')[1])
            area = math.pi * r ** 2
        else:
            continue

        print(f"{area:.2f}")

computeAreas("TH13-16.txt")
```

```python
#W11A19
import struct

def read_binary_doubles():
    file_path = input().strip()
    
    try:

        with open(file_path, 'rb') as f:
            data = f.read()
            double_size = 8
            if len(data) % double_size != 0:
                print("Cảnh báo: Tệp có thể không chỉ chứa các số double hoặc bị thiếu dữ liệu.")
            num_elements = len(data) // double_size
            format_string = f'{num_elements}d'
            doubles = struct.unpack(format_string, data)

            for value in doubles:
                print(f"{value:.5f}")
                
    except FileNotFoundError:
        print("Lỗi: Không tìm thấy tệp tin.")
    except Exception as e:
        print(f"Đã xảy ra lỗi: {e}")

if __name__ == "__main__":
    read_binary_doubles()
```

```python
#W11A18
import struct

def read_binary_doubles():
    file_path = input().strip()
    
    try:
        with open(file_path, 'rb') as f:
            data = f.read()
            byte_size = 8
            num_elements = len(data) // byte_size
            
            if num_elements == 0:
                print("Tệp rỗng hoặc không đủ dữ liệu.")
                return

            format_string = f'{num_elements}d'
            doubles = struct.unpack(format_string, data)
            for value in doubles:
                print(f"{value:.5f}")
                
    except FileNotFoundError:
        print(f"Lỗi: Không tìm thấy tệp tại đường dẫn: {file_path}")
    except struct.error:
        print("Lỗi: Định dạng tệp không hợp lệ (không phải danh sách các số double).")
    except Exception as e:
        print(f"Đã xảy ra lỗi: {e}")

if __name__ == "__main__":
    read_binary_doubles()
```

```python
#W11A17
import math
import re

def estimatedTime(path: str):
    try:
        with open(path, 'r', encoding='utf-8') as f:
            content = f.read()

        coords = re.findall(r"[-+]?\d*\.\d+|\d+", content)
        
        x1, y1 = float(coords[0]), float(coords[1])
        x2, y2 = float(coords[2]), float(coords[3])
        v = float(coords[4])

        distance = math.sqrt((x2 - x1)**2 + (y2 - y1)**2)
        
        time_seconds = distance / v
        
        return round(time_seconds, 2)

    except FileNotFoundError:
        return "Lỗi: Không tìm thấy tệp."
    except Exception as e:
        return f"Đã xảy ra lỗi: {e}"
```
