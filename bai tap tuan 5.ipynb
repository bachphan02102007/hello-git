```python
#W5A2
def swap(a,b):
    a,b=b,a
    return a,b
a,b=map(int,input().split())
print(swap(a,b))
```

```python
#W5A3
import math
def is_prime(a):
    for i in range(2,math.isqrt(a)+1):
        if a%i==0:
            return 'k phai so nto'
    else:
        return 'la so nto'
a=int(input())
print(is_prime(a))
        
```

```python
#W5A4
def perfect_number(a):
    tong=0
    for i in range(1,a//2+1):
        if a%i==0:
            tong+=i
    if tong==a:
        return'is perfect'
    else:
        return'not perfect'
a=int(input())
print(perfect_number(a))
```

```python
#W5A5
def vi_tri(a,b):
    if b in a:
        vt=a.index(b)+1
        return vt
    else :
        return -1
a=list(map(int,input().split()))
b=int(input())
print(vi_tri(a,b))
```

```python
#W5A6
def gt(a):
    b=1
    for i in range(1,a+1):
        b*=i
    return b
a=int(input())
print(gt(a))
```

```python
#W5A7
def may_tinh_mini(so1,so2,dau):
    if dau=='+':
        kq=so1 + so2
    elif dau=='-':
        kq=so1-so2
    elif dau=='*':
        kq=so1*so2
    elif dau=='/':
        if so2==0:
            return 'k the thuc hien phep tinh'
        else :
            kq=so1/so2
    return round(kq,2)
so1_str,dau,so2_str=input().split()
so1=float(so1_str)
so2=float(so2_str)
print(may_tinh_mini(so1,so2,dau))


```

```python
#W5A8
def hamming(a,b):
    c=[]
    d=[]
    while a>0:
        c.append(a%2)
        a=a//2
    while b>0:
        d.append(a%2)
        b=b//2
    dem=0
    if len(c)>=len(d):
        d=d+[0]*(len(c)-len(d))
        for i,u in zip(c,d):
            if i!=u:
                dem+=1
    else :
        c=c+[0]*(len(d)-len(c))
        for i,u in zip(c,d):
            if i!=u:
                dem+=1
    return dem
a,b=map(int,input().split())
print(hamming(a,b))
```

```python
#W5A9
def tong_chu_so(a):
    tong=0
    while a>0:
        tong+=a%10
        a//=10
    return tong
a=int(input())
print(tong_chu_so(a))
```

```python
#W5A10
def anh_xa(a,b):
    for i in range(len(a)):
        if a.index(a[i])!=b.index(b[i]):
            return False
    return True
a,b=input().split()
print(anh_xa(a,b))
```
